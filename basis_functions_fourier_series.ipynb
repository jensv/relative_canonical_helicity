```python
import numpy as np
import scipy.fftpack as fft

%matplotlib inline
import matplotlib.pyplot as plt

import seaborn as sns
sns.set_style('white')
sns.set_context('poster')
```

```python
n = 30
x = np.linspace(np.pi/2, 3*np.pi/2, n)
dx = x[1] - x[0]

y = np.sin(x)

#y = -(x-4)**4-10*(x-3)**3+50*(x)**2+4
plt.plot(x, y)
```

```python
fig, axes = plt.subplots(1,2)

fft_y = fft.fft(y)
freq = fft.fftfreq(n, dx)
k_non_period = freq
axes[0].plot(freq[0:n/2], np.abs(fft_y)[0:n/2], label='fft')
axes[1].plot(freq[0:n/2], np.abs(fft_y)[0:n/2], label='fft')

fft_y = fft.dst(y)
freq = fft.fftfreq(n, dx)
axes[0].plot(freq[0:n/2], np.abs(fft_y)[0:n/2], label='dst')
axes[1].plot(freq[0:n/2], np.abs(fft_y)[0:n/2], label='dst')

fft_y = fft.dct(y)
freq = fft.fftfreq(n, dx)
axes[0].plot(freq[0:n/2], np.abs(fft_y)[0:n/2], label='dct')
axes[1].plot(freq[0:n/2], np.abs(fft_y)[0:n/2], label='dct')

plt.yscale('log')
plt.legend()
plt.show()
```
