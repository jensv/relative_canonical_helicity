# Imports

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context('poster')
sns.set_style('whitegrid')
from scipy.optimize import leastsq
from scipy.optimize import curve_fit

from mpl_toolkits.mplot3d import Axes3D
```

# Look at $\Omega_{i raw}$ Streamlines from time point 004

# Load Data

```python
input_path = "../input/2016-10-29/"
B_center = np.loadtxt(input_path +
                      "B_center_004.xyz", 
                      skiprows=2,
                      usecols=(1, 2, 3))
Omega_i_center = np.loadtxt(input_path +
                            "Omega_i_raw_plus_center_004.xyz", 
                            skiprows=2,
                            usecols=(1, 2, 3))

angles = [0, 90, 180, 270]
Omega_i_inner = []
Omega_i_outer = []

for i in range(4):
    file_name = (input_path + "Omega_i_raw_plus_inner_" + str(angles[i]) + "_004.xyz")
    Omega_i_inner.append(np.loadtxt(file_name, skiprows=2, usecols=(1, 2, 3)))

for i in range(4):
    file_name = (input_path + "Omega_i_raw_plus_outer_" + str(angles[i]) + "_004.xyz")
    Omega_i_outer.append(np.loadtxt(file_name, skiprows=2, usecols=(1, 2, 3)))
```

# Plot x,y data

```python
fig = plt.figure()
axis = fig.add_subplot(111, projection='3d')

axis.scatter(B_center[:, 0], B_center[:, 1], 
             B_center[:, 2], 
             color='red', s=50, marker='.')
axis.scatter(Omega_i_center[:, 0], Omega_i_center[:, 1],
             Omega_i_center[:, 2],
             color='black', s=50, marker='o')
for inner in Omega_i_inner:
    axis.scatter(inner[:, 0], inner[:, 1],
                 inner[:, 2],
                 s=50, marker='o', color='grey')
for outer in Omega_i_outer: 
    axis.scatter(outer[:, 0], outer[:, 1],
                outer[:, 2],
                s=50, marker='o', color='grey', alpha = 0.5)
```

```python
fig = plt.figure()
axis = fig.add_subplot(111, projection='3d')
axis.scatter(Omega_i_center[:, 0], Omega_i_center[:, 1],
             Omega_i_center[:, 2],
             color='black', s=50, marker='o')
```

```python
fig = plt.figure()
axis = fig.add_subplot(111, projection='3d')

axis.scatter(Omega_i_center[:, 0], Omega_i_center[:, 1],
             Omega_i_center[:, 2],
             color='black', s=50, marker='o')

points_z = np.linspace(0.249, 0.41, 100)
fit_x = line(points_z, x_slope, x_offset)
fit_y = line(points_z, y_slope, y_offset)
axis.plot(fit_x, fit_y, points_z)
```
